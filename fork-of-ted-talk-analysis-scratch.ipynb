## Average Views of Bill Gates Ted Talks

```python
view = []
for i  in range(len(data)):
    if(data[i][1] == 'Bill Gates'):
        view.append(data[i][3])
int(sum(view)/len(view))
```

## View to like Ratio

```python
view_to_like = []
for i  in range(len(data)):
    view_to_like.append(int(data[i][3]/data[i][4]))
view_to_like
```

```python
min(view_to_like)
```

```python
max(view_to_like)
```

## How Many TED Talks were Given in 2020? 

```python
data = df.values

month = []
year = []

for i in range(len(data)):
    month.append(data[i][2].split(' ')[0])
    year.append(data[i][2].split(' ')[1])
```

```python
df['year'] = year
df['month'] = month
```

```python
df.head()
```

```python
len(df[df['year'] == '2020'])
```
